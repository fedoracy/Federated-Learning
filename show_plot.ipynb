```python
import json
import matplotlib.pyplot as plt

def plot_stats(test_train_valid_loss_acc):
    fPath = './stats.txt'
    f = open(fPath, 'r') 
    dic = json.loads([i for i in f.readlines()][0])
    f.close()
    round_list,time_list,y_list = [], [], []
    for (r,t,l) in dic[test_train_valid_loss_acc]:
        round_list.append(r)
        time_list.append(t)
        y_list.append(l)
    fig = plt.figure() 
    plt.xlabel('X')  
    plt.ylabel('Y')  
    plt.plot(round_list,y_list,c = 'r',marker = 'o')  
    plt.legend('x1')  
    plt.show()  
```

```python
for i in ['train_loss',
          'valid_loss',
          'test_loss',
          'train_accuracy',
          'valid_accuracy', 
          'test_accuracy']:
    print(i)
    plot_stats(i)
```
